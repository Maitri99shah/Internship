1. Scrape the details of most viewed videos on YouTube from Wikipedia. Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos You need to find following details: 
A) Rank
B) Name
C) Artist
D) Upload date
E) Views

```python
driver = webdriver.Chrome()
```

```python
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
```

```python
//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[1]/sup/a

```

```python
rank = []
name = []
artist = []
upload_date = []
views = []
```

```python
rank_tags = driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[1]/sup/a ')
for i in rank_tags:
    num = i.text
    rank.append(num)
```

```python
i = 1
n = 30
for i in range(i,n+1):
    rank.append(i)
    i+=1
```

```python
name_tags = driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[1]')
for i in name_tags:
    try:
        names = i.text
        name.append(names)
    except NoSuchElementException:
        name.append('--')
```

```python
identity = [i.split('[', 1)[0] for i in name]
```

```python
artist_tags = driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[2]')
for i in artist_tags:
    try:
        aname = i.text
        artist.append(aname)
    except NoSuchElementException:
        artist.append('--')
```

```python
upload_date_tags = driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[4]')
for i in upload_date_tags:
    try:
        udate = i.text
        upload_date.append(udate)
    except NoSuchElementException:
        upload_date.append('--')
```

```python
views_tags = driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[3]')
for i in views_tags:
    try:
        view = i.text
        views.append(view)
    except NoSuchElementException:
        views.append('--')
```

```python
print(len(rank),len(identity),len(artist),len(upload_date),len(views))
```

```python
df = pd.DataFrame({'Rank ':rank,'Name ':identity,'Artist ':artist,'Upload Date ':upload_date,'Views (in Billion) ':views})
df
```
