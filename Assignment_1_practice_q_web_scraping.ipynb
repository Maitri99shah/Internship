3. Write a python program to scrape mentioned details from dineout.co.in : 
    i) Restaurant name 
    ii) Cuisine
    iii) Location
    iv) Ratings 
    v) Image URL.

```python
page = requests.get('https://www.dineout.co.in/delhi-restaurants/value-for-money')
page
```

```python
soup = BeautifulSoup(page.content)
soup
```

```python
restaurant_name =[]
for i in soup.find_all('div',class_="restnt-info cursor"):
    restaurant_name.append(i.text)
    
restaurant_name
```

```python
price_cuisine =[]

for i in soup.find_all('span',class_="double-line-ellipsis"):
    price_cuisine.append(i.text.replace)

price_cuisine
```

```python
price = price_cuisine.copy()
price
```

```python
price = [i.split('|', 1)[0] for i in price_cuisine]
price
```

```python
cuisine = [i.split('|', 1)[1] for i in price_cuisine]
cuisine
```

```python
location =[]

for i in soup.find_all('div',class_="restnt-loc ellipsis"):
    location.append(i.text)
    
location
```

```python
rating3 =[]
rating4 =[]
rating5 =[]
for i in soup.find_all('div',class_="restnt-rating rating-3"):
    rating3.append(i.text)
    
for i in soup.find_all('div',class_="restnt-rating rating-4"):
    rating4.append(i.text)
    
for i in soup.find_all('div',class_="restnt-rating rating-5"):
    rating5.append(i.text)

rating3_5 = rating3+ rating5
pos = 8
for i in range(len(rating3_5)):
    rating4.insert(i + pos, rating3_5[i])
    
rating4
```

```python
image_url = []

for i in soup.find_all('img',class_="no-img"):
    image_url.append(i['data-src'])
    
image_url
```

```python
print(len(restaurant_name),len(cuisine),len(location),len(rating4),len(image_url))
```

```python
import pandas as pd
```

```python
final= pd.DataFrame({'Restaurant Name ':restaurant_name,'Cuisine ':cuisine,'Location ':location,'Ratings ':rating4,'Image URLs ':image_url})
final
```
