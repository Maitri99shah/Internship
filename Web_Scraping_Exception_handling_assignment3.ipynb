**3.Write a python program to access the search bar and search button on images.google.com and scrape 10 images each for keywords ‘fruits’, ‘cars’ and ‘Machine Learning’, ‘Guitar’, ‘Cakes’.

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Set the path to the web driver executable
driver_path = 'path_to_chromedriver'

# Create a new instance of the Chrome driver
driver = webdriver.Chrome(driver_path)

# Open images.google.com
driver.get('https://images.google.com')

# Find the search bar element and enter the keywords
search_bar = driver.find_element(By.CLASS_NAME, 'Tg7LZd')
keywords = ['fruits', 'cars', 'Machine Learning', 'Guitar', 'Cakes']

for keyword in keywords:
    search_bar.clear()
    search_bar.send_keys(keyword)
    search_bar.send_keys(Keys.RETURN)

  # Wait for the search results to load
    WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME, 'YQ4gaf')))

  # Scrape the image URLs
    image_elements = driver.find_elements(By.CLASS_NAME, 'YQ4gaf')
    image_urls = [element.get_attribute('src') for element in image_elements]

  # Print the first 10 image URLs
    print(f"Top 10 images for '{keyword}':")
    for url in image_urls[:10]:
        print(url)
```

**9.Write a python program to scrape a data for all available Hostels from https://www.hostelworld.com/ in “London” location. You have to scrape hostel name, distance from city centre, ratings, total reviews, overall reviews, privates from price, dorms from price, facilities and property description.

```python
driver = webdriver.Chrome()
```

```python
driver.get("https://www.hostelworld.com/")
```

```python
place = driver.find_element(By.XPATH,"/html/body/div[3]/div/div[2]/main/header/div/div[2]/div[1]/div[1]/div/div[1]/div[1]/div/div[2]/input")
place.send_keys('london')
```

```python
found_place = driver.find_element(By.XPATH,"/html/body/div[3]/div/div[2]/main/header/div/div[2]/div[1]/div[1]/div/div[1]/div[2]/div/ul/li[2]/button/div[2]/div[1]")
found_place.click()
```

```python
search = driver.find_element(By.XPATH,"/html/body/div[3]/div/div[2]/main/header/div/div[2]/div[1]/div[1]/div/div[5]/button[1]")
search.click()
```

```python
hostel_name = []
distance = []
ratings = []
total_reviews = []
overall_reviews = []
private_from_price = []
dorms_from_price = []
facilities = []
property_description = []
```

```python
hostel_name_tags = driver.find_elements(By.XPATH,'//div[@class="property-name"]')
for i in hostel_name_tags:
    try:
        names = i.text
        hostel_name.append(names)
    except NoSuchElementException:
        hostel_name.append('--')
```

```python
distance_tags = driver.find_elements(By.XPATH,'//span[@class="distance-description"]')
for i in distance_tags:
    try:
        distances = i.text
        distance.append(distances)
    except NoSuchElementException:
        distance.append('--')
```

```python
ratings_tags = driver.find_elements(By.XPATH,'//div[@class="property-info-header"]/div/div[1]/div[1]/span[1]')
for i in ratings_tags:
    try:
        rating = i.text
        ratings.append(rating)
    except NoSuchElementException:
        ratings.append('--')
```

```python
total_reviews_tags = driver.find_elements(By.XPATH,'//div[@class="property-info-header"]/div/div[1]/div[2]')
for i in total_reviews_tags:
    try:
        total_review = i.text
        total_reviews.append(total_review)
    except NoSuchElementException:
        total_reviews.append('--')
```

```python
overall_reviews_tags = driver.find_elements(By.XPATH,'//div[@class="property-info-header"]/div/div[1]/div[1]/span[2]')
for i in overall_reviews_tags:
    try:
        overall_review = i.text
        overall_reviews.append(overall_review)
    except NoSuchElementException:
        overall_reviews.append('--')
```

```python
urls = []
urls_tags = driver.find_elements(By.XPATH,'//a[@class="property-card-container horizontal"]')
for i in urls_tags:
    url = i.get_attribute('href')
    urls.append(url)
```

```python
urls
```

```python
for i in urls:
    driver.get(i)
    time.sleep(0.5)
    
    try:
        private_from_price_tags = driver.find_element(By.XPATH,'/html/body/div[2]/div/div[5]/div[3]/div[4]/div[49]/div/table/tbody/tr[1]/td[2]/span')
        private_from_price.append(private_from_price_tags.text)   
    except NoSuchElementException:
        private_from_price.append('--')
```

```python
for i in urls:
    driver.get(i)
    time.sleep(0.5)
    
    try:
        dorms_from_price_tags = driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/section/div[12]/div/div/section/div/div[1]/div/div[2]/div[2]/div[1]/div[2]/div[1]/div/div[1]/div[2]/div[1]')
        dorms_from_price.append(dorms_from_price_tags.text)   
    except NoSuchElementException:
        dorms_from_price.append('--')
```

```python
for i in urls:
    driver.get(i)
    time.sleep(0.5)
    
    try:
        facilities_tags = driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/section/div[14]/div/ul/li[1]/ul')
        facilities.append(facilities_tags.text)   
    except NoSuchElementException:
        facilities.append('--')
```

```python
for i in urls:
    driver.get(i)
    time.sleep(0.5)
    
    try:
        property_description_tags = driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/section/div[10]/div/div[2]/div[6]/div/div[2]')
        property_description.append(property_description_tags.text)   
    except NoSuchElementException:
        property_description.append('--')
```

```python
df = pd.DataFrame({'Hostel Name ':hostel_name,'Distance From City Centre ':distance,'Ratings ':ratings,'Total Reviews ':total_reviews,'Overall Reviews ':overall_reviews,'Private From Price ':private_from_price,'Dorm From Price ':dorms_from_price,'Facilities ':facilities,'Property Description':property_description})
df
```
