**Q1: In this question you have to scrape data using the filters available on the webpage You have to use the location and salary filter.

**Q3: Scrape 100 reviews data from flipkart.com for iphone11 phone. You have to go the link: 
https://www.flipkart.com/apple-iphone-11-black-64-gb/product-reviews/itm4e5041ba101fd?pid=MOBFWQ6BXGJCEYNY&lid=LSTMOBFWQ6BXGJCEYNYZXSHRJ&marketplace=FLIPKART**

```python
driver = webdriver.Chrome()
```

```python
driver.get("https://www.flipkart.com/apple-iphone-11-black-64-gb/product-reviews/itm4e5041ba101fd?pid=MOBFWQ6BXGJCEYNY&lid=LSTMOBFWQ6BXGJCEYNYZXSHRJ&marketplace=FLIPKART")
```

```python
ratings = []
review_summary = []
full_reviews = []
```

```python
start = 0
end = 10
for page in range(start,end):
    
    rating_tags = driver.find_elements(By.XPATH,'//div[@class="_3LWZlK _1BLPMq"]')
    for i in rating_tags:
        rating = i.text
        ratings.append(rating)
    
    review_tags = driver.find_elements(By.XPATH,'//p[@class="_2-N8zT"]')
    for i in review_tags:
        review = i.text
        review_summary.append(review)
    
    full_review_tags = driver.find_elements(By.XPATH,'//div[@class="t-ZTKy"]/div/div')
    for i in full_review_tags:
        full_review = i.text
        full_reviews.append(full_review)
    
    next_button = driver.find_element(By.XPATH,'/html/body/div/div/div[3]/div/div/div[2]/div[13]/div/div/nav/a[11]')
    next_button.click()
    time.sleep(10)
```

```python
print(len(ratings),len(review_summary),len(full_reviews))
```

**Q6: Write a python program to scrape data for Top 1000 Quotes of All Time.
The above task will be done in following steps:
1. First get the webpage https://www.azquotes.com/
2. Click on Top Quote
3. Than scrap a) Quote b) Author c) Type Of Quotes

```python
driver = webdriver.Chrome()
```

```python
driver.get("https://www.azquotes.com/")
```

```python
top_quote = driver.find_element(By.XPATH,'//div[@class="mainmenu"]/ul/li[5]/a')
top_quote.click()
```

```python
quotes = []
author = []
type_quotes = []
```

```python
start = 0
end = 10
for page in range(start,end):
    
    quote_tags = driver.find_elements(By.XPATH,'//a[@class="title"]')
    for i in quote_tags:
        quote = i.text
        quotes.append(quote)
        
    author_tags = driver.find_elements(By.XPATH,'//div[@class="author"]/a')
    for i in author_tags:
        authors = i.text
        author.append(authors)
        
    type_quotes_tags = driver.find_elements(By.XPATH,'//div[@class="tags"]')
    for i in type_quotes_tags:
        quote_types = i.text
        type_quotes.append(quote_types)
    
    next_button = driver.find_element(By.XPATH,'/html/body/div[1]/div[2]/div/div/div/div[1]/div/div[4]/li[12]/a')
    next_button.click()
    time.sleep(10)
    
```

```python
print(len(quotes),len(author),len(type_quotes))
```

```python
df = pd.DataFrame({'Quote ':quotes,'Author ':author,'Type Of Quotes ':type_quotes})
```

**Q7: Write a python program to display list of respected former Prime Ministers of India (i.e. Name, Born-Dead, Term of office, Remarks) from https://www.jagranjosh.com/general-knowledge/list-of-all-prime-ministers-of-india-1473165149-1
scrap the mentioned data and make the DataFrame

```python
driver = webdriver.Chrome()
```

```python
driver.get("https://www.jagranjosh.com/general-knowledge/list-of-all-prime-ministers-of-india-1473165149-1")
```

```python
name = []
born_dead = []
term_of_office = []
remarks = []
```

```python
name_tags = driver.find_elements(By.XPATH,'//div[@class="Details_StoryBody__85cfI"]/div[6]/div/table[1]/tbody/tr/td[2]')
for i in name_tags:
    names = i.text
    name.append(names)
        
born_dead_tags = driver.find_elements(By.XPATH,'//div[@class="Details_StoryBody__85cfI"]/div[6]/div/table[1]/tbody/tr/td[3]')
for i in born_dead_tags:
    years = i.text
    born_dead.append(years)
        
term_of_office_tags = driver.find_elements(By.XPATH,'//div[@class="Details_StoryBody__85cfI"]/div[6]/div/table[1]/tbody/tr/td[4]')
for i in term_of_office_tags:
    period = i.text
    term_of_office.append(period)
        
remark_tags = driver.find_elements(By.XPATH,'//div[@class="Details_StoryBody__85cfI"]/div[6]/div/table[1]/tbody/tr/td[5]')
for i in remark_tags:
    content = i.text
    remarks.append(content)
```

```python
print(len(name),len(born_dead),len(term_of_office),len(remarks))
```

```python
df = pd.DataFrame({'Name ':name, 'Born-Dead ':born_dead, 'Term of office ':term_of_office, 'Remark ':remarks})
df
```

**Q8: Write a python program to display list of 50 Most expensive cars in the world (i.e. Car name and Price) from https://www.motor1.com/
This task will be done in following steps:
1. First get the webpage https://www.motor1.com/
2. Then You have to type in the search bar ‚Äô50 most expensive cars‚Äô
3. Then click on 50 most expensive cars in the world..
4. Then scrap the mentioned data and make the dataframe.

```python
driver = webdriver.Chrome()
```

```python
driver.get("https://www.motor1.com/")
```

```python
search_cars = driver.find_element(By.XPATH,'/html/body/div[9]/div[2]/div/div/div[3]/div/div/div/form/input')
search_cars.send_keys('50 most expensive cars in the world')
```

```python
search = driver.find_element(By.XPATH,'/html/body/div[9]/div[2]/div/div/div[3]/div/div/div/form/button[1]')
search.click()
```

```python
search_card = driver.find_element(By.XPATH,'/html/body/div[9]/div[9]/div/div[1]/div/div/div[1]/div')
search_card.click()
```

```python
name = []
price = []
img_url = []
```

```python
name_tags = driver.find_elements(By.XPATH,'//h3[@class="subheader"]')
for i in name_tags:
    car_name = i.text
    name.append(car_name)
```

```python
price_tags = driver.find_elements(By.XPATH,'//div[@class="postBody description e-content"]/p/strong')
for i in price_tags:
    cost = i.text
    price.append(cost)
```

```python
img_url_tags= driver.find_elements(By.XPATH,'/html/body/div[9]/div[7]/div[2]/div[1]/div[2]/div[2]/section/picture/img')
for i in img_url_tags:
    url = i.get_attribute('src')
    img_url.append(url)
```

```python
print(len(name),len(price),len(img_url))
```

```python
df = pd.DataFrame({'Car Name ':name, 'Car Price':price, 'Image_URL':img_url})
df
```
